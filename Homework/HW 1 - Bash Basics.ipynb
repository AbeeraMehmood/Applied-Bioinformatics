### A1.ab: Execute basic bash commands in R (using Jupyter) [20-30 min] TPS
#### ab.i Confirmation of basic bash operations <br>
Usually, you would want to use command line in your terminal to run an R script for a given job/task. Here we are going to leverage the `%%bash` magic command in the jupyter python kernel to execute bash commands in the notebok directly. More information on mafic commands can be found in the [documentation of the python kernel](https://ipython.readthedocs.io/en/stable/interactive/magics.html). If you want to call bash commands using the R kernel, you will utilize the [`system`](https://www.rdocumentation.org/packages/base/versions/3.6.2/topics/system) function in R.

For example to execute the `pwd` command in the python kernel, you will use, 

```
%%bash
pwd
```

To execute the same `pwd` command in the R kernel, you will use, 

```
system('pwd', intern=TRUE)
```

In this notebook, we will **only be using the python kernel**. You can change the kernel that your notebook by clicking on `Kernel > Change Kernel` in the menu bar. <img src="./img/kernel_change.png" width="400" />
